```python
# Adicionar 1 coluna
# de uma coluna existente
vendas_df['Comissão'] = vendas_df['Valor Final']*0.05
display(vendas_df)
```

```python
# criar coluna nova
vendas_df.loc[:, 'Imposto'] = 0
display(vendas_df)
```

```python
# Excluir linhas e colunas
# 1 = coluna ; 0 = linha

vendas_df = vendas_df.drop('Imposto', axis=1)
```

```python
display(vendas_df)
```

```python
vendas_df = vendas_df.drop(0, axis=0)
```

```python
display(vendas_df)
```

```python
# deletar linhas e colunas completamente vazias
# all = excluir linhas e colunas completamente vazias
# 1 = coluna ; 0 = linha
#vendas_df = vendas_df.dropna(how='all', axis=1)

# deletar linhas com pelo menos 1 vazio
#vendas_df = vendas_df.dropna() 

# Preencher valores vazios
# ta preenchendo com a média dos valores de comissão
vendas_df['Comissão'] = vendas_df['Comissão'].fillna(vendas_df['Comissão'].mean)
display(vendas_df)
```

```python
# Preencher com valor de cima
vendas_df = vendas_df.fill()
```

```python
# Calcular indicadores (Groupby, Value count)

# Quantas vendas eu tive?
transacoes_loja = vendas_df['ID Loja'].value_counts()
print(transacoes_loja)
```

```python
# faturamento dos produtos - agrupar somando
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_produto)
```

```python
# Média dos produtos - agrupar média
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').mean()
display(faturamento_produto)
```

```python
# importar tabela gerentes
gerentes_df = pd.read_excel('Gerentes.xlsx')
display(gerentes_df)
```

```python
# adicionar coluna gerente
vendas_df = vendas_df.merge(gerentes_df)
display(vendas_df)
```
