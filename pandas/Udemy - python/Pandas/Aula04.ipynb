```python
# MERGE E GROUPBY
import pandas as pd
```

```python
# Cadastro Loja a
cadastro_a = {"Id": ["AA2930", "BB4563", "CC2139", "DE2521", "GT3462", "HH1158"],
              "Nome": ["Victor", "Amanda", "Bruna", "Carlos", "Ricardo", "Maria"],
              "Idade": [20,35,40,54,30,27],
              "CEP": ["00092-029", "11111-111", "22222-888", "00000-999", "88888-111", "77777-666"]
}
cadastro_a = pd.DataFrame(cadastro_a, columns=["Id","Nome","Idade","CEP"])
cadastro_a
```

```python
# Cadastro Loja B
cadastro_b = {"Id": ["CC9999", "EF4488", "DD9999", "GT3462","HH1158"],
              "Nome": ["Marcos", "Patricia", "Erika", "Ricardo", "Maria"],
              "Idade": [19,30,22,30,27],
              "CEP": ["00092-029", "11111-111", "11111-888", "88888-111", "77777-666"]
}
cadastro_b = pd.DataFrame(cadastro_b, columns=["Id","Nome","Idade","CEP"])
cadastro_b
```

```python
# Registro de compras
compras = {"Id": ["AA2930","EF4488","CC2139","EF4488","CC9999","AA2930","HH1158","HH1158"],
              "Data": ["2019-01-01","2019-01-30","2019-01-30","2019-02-01","2019-02-20","2019-03-15","2019-04-08", "2019-04-25" ],
              "Valor": [200,100,40,150,300,25,50,500]
}
compras = pd.DataFrame(compras, columns=["Id","Data","Valor"])
compras
```

```python
pd.merge(cadastro_a, cadastro_b, on = ["Id"], how = "inner") # clientes da loja A que tambem estão na B
```

```python
pd.merge(cadastro_a, cadastro_b[["Id","Idade","CEP"]], on=["Id"], how = "inner") # só mostra ID, IDADE e CEP do B e Tudo do A
```

```python
pd.merge(cadastro_a, cadastro_b[["Id","Idade","CEP"]], on=["Id"], how = "inner", suffixes=("_A","_B")) # Mudar a nomeclatura
```

```python
lojas = pd.concat([cadastro_a,cadastro_b],  ignore_index=True) # full join - porem com duplicadas
```

```python
lojas
```

```python
clientes_unicos = lojas.drop_duplicates(subset="Id") # Remover duplicatas
```

```python
esquerda = pd.merge(cadastro_a, compras, how = "left", on = ["Id"]) # colocar valores de compras na A
```

```python
esquerda.groupby(["Id", "Nome"])["Valor"].sum() # soma os valores de cada Id, deixando o nome
```

```python
pd.merge(cadastro_a, cadastro_b, on = ["Id"], how = "outer") # junta todos os dados, mas sem unir as colunas
```

```python
pd.merge(cadastro_a, cadastro_b, on = ["Id"], how = "outer", indicator=True) # vai indicar quando tem dados em uma e na outra
```
