```python
# Ordenação dos dados

import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({"Col1":["A","A","B", np.nan,"D","C"],"Col2":[2,1,8,9,7,4],"Col3":[0,1,4,3,6,5]})
```

```python
df
```

```python
df.sort_values(by = "Col1") # Crescente alfabética
```

```python
df.sort_values(by = "Col2") # Crescente 
```

```python
df.sort_values(by =[ "Col1", "Col2"]) # Crescente multi colunas
```

```python
df.sort_values(by = "Col1", ascending= False) # Descrescente
```
